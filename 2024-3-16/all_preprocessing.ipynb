```python
from konlpy.tag import Kkma
```

```python
kkma = Kkma()
```

```python
target_words = [
    "연필", "자", "가위", "망치", "톱", "칫솔", "빗", "숟가락", "국자", "칼",
    "사과", "주스", "빵", "바지", "양말", "모자", "장갑", "선풍기", "편지", "달력", "의자",
    "고양이", "사자", "이모", "삼촌", "누나", "동생", "소방관", "가수"
] # 말 시트 수정 필요
```

```python
import pandas as pd
```

```python
# 파일명
xls = pd.ExcelFile('penciltree_old data.xlsx')

# Daraframe형식으로 엑셀 파일 읽기
df = pd.read_excel(xls, engine='openpyxl', sheet_name=target_words)
```

```python
tagset = ["EFN", "ETD", "JKM", "NNG", "SF", "VV", "ECS", "EPT", "JKO", "XSV", "ECD", "JKS", "MAG", "VXV", "EFQ", "JX", "VA", "XSA", "ETN", "NNB", "VCP", "JC", "NP", "ECE", "XSN", "VX", "VXA", "SP", "MDN", "NNM", "NR", "EFA", "SS", "JKG", "OL", "XR", "EPH", "MDT", "NNP"]
```

```python
from collections import defaultdict


def get_columns(tagset, texts):
    columns = {tag: list() for tag in tagset}

    for text in texts:
        tag_dict = defaultdict(list)
        morphs = kkma.pos(text)
        for morph, tag in morphs:
            tag_dict[tag].append(morph)

        for tag in columns.keys():
            if vals := tag_dict.get(tag):
                columns[tag].append(",".join(vals))
            else:
              columns[tag].append(None)

    return columns
```

```python
for word in target_words:
    print(f"{word} sheet processing...")
    texts = df[word][word]
    columns = get_columns(tagset, texts)
    for tag, value in columns.items():
        df[word][tag] = value
print("Proccess Completed")
```

```python
# 엑셀 생성
import os

dir_name = "result_all"
file_name = "old_result_all"

if not os.path.exists(dir_name):
    os.mkdir(dir_name)

for word in target_words:
    pd.DataFrame(df[word]).to_excel(f'{dir_name}/{word}_{file_name}.xlsx', index=False)
```
